```python
train = pd.read_csv("/kaggle/input/titanic/train.csv",index_col=[0])
```

```python
train.Age = train.Age.fillna(train.Age.mean())
```

```python
train.nunique()
```

```python
op = train[['Survived']]
train.drop(['Cabin','Survived','Name','Ticket'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
def preprocess(data):
    le = LabelEncoder()
    data.Sex = le.fit_transform(data.Sex)
    data.Embarked = le.fit_transform(data.Embarked)
    return data
```

```python
train_proce = preprocess(train)
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
model = DecisionTreeRegressor(random_state=1)

# Fit model
model.fit(train_proce,op)
```

```python
test = pd.read_csv('/kaggle/input/titanic/test.csv',index_col=[0])
```

```python
test.Fare = test.Fare.fillna(test.Fare.mean())
```

```python
test.Age = test.Age.fillna(train.Age.mean())
test.drop(['Cabin','Name','Ticket'],axis=1,inplace=True)
test = preprocess(test)
ans = model.predict(test)
pd.DataFrame(ans,index = test.index,columns = ['Survived']).to_csv("test_res.csv")
```
