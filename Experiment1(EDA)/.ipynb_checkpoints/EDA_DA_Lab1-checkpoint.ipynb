```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/data/dataset/train.csv',index_col=[0])
data.head()
```

```python
data.info()
```

In the above command we can say that there are 14 columns in the data set. The types of the data present are object, float and integer. Bty closely oberserving the non null count we can also infer that there are also null values in the dataset

```python
# Here using this command we will see the number of the 
data.isnull().sum()
```

```python
print("The data size with null values",data.shape)
data.dropna(inplace=True) # this command is use to remove the null values of the dataset
print("The data sizen after dropping null values",data.shape)
```

```python
data.isnull().sum()
```

```python
print(data.nunique(),"\n") # getting the number unique values for each columns
print("Propety Type",data.Property_Type.unique().tolist())
print("Power Backup",data.Power_Backup.unique().tolist())
print("Water Supply",data.Water_Supply.unique().tolist())
print("Furnishing",data.Furnishing.unique().tolist())
print("CrimeRate",data.Crime_Rate.unique().tolist())
print("Dust and Noise",data.Dust_and_Noise.unique().tolist())
```

Till now we have explored the data columns there data variety. We have also remove the null values. Now we will try to explore the data set using some grpahs. We have also explored the unique values in the data set

```python
# this command will give us the 5 point summary
data.describe()
```

Now we will do the data visualization

```python
# in this we are plotting the distribution of the Habitability score and Property area
# THis is just to understand how the these two data spread around the mean and verify the results
# from the 5 point summary
plt.figure(figsize=(9, 8))
sns.distplot(data['Habitability_score'], color='g', bins=100, hist_kws={'alpha': 0.4})
plt.figure(figsize=(9, 8))
sns.distplot(data['Property_Area'], color='g', bins=100, hist_kws={'alpha': 0.4})
```

```python
plt.figure(figsize=(10, 10))
palette_color = sns.color_palette('bwr')
plt.pie(data.groupby('Property_Type').size(),labels = data.groupby('Property_Type').size().index, colors=palette_color, autopct='%.0f%%',textprops={'fontsize': 15})
plt.title('Property_Type',fontsize=20)

plt.figure(figsize=(10, 10))
plt.pie(data.groupby('Crime_Rate').size(),labels = data.groupby('Crime_Rate').size().index, colors=palette_color, autopct='%.0f%%',textprops={'fontsize': 15})
plt.title('Crime_Rate',fontsize=20)

plt.figure(figsize=(10, 10))
plt.pie(data.groupby('Furnishing').size(),labels = data.groupby('Furnishing').size().index, colors=palette_color, autopct='%.0f%%',textprops={'fontsize': 15})
plt.title('Furnishing',fontsize=20)
```

```python
# We will be selecting only some of the numberical features of the dataset
df_num = data.loc[:,['Property_Area','Number_of_Windows',"Number_of_Doors",'Traffic_Density_Score','Air_Quality_Index','Neighborhood_Review','Habitability_score']]
# df_num = data.select_dtypes(include = 'number')
df_num.head()
```

```python
corr = df_num.drop('Habitability_score', axis=1).corr() 
# We find the correlation of the various feature to find the relation among the other features

plt.figure(figsize=(12, 10))
# Here we will be plotting the heatmap of the correalation matrix obtaines
sns.heatmap(corr, 
            cmap='Accent', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True)
```

From above we can infer that non of the selected fratures show correlation with each other. We will confirm this inference usng scatterplot between the features.

```python
sns.pairplot(df_num)
```

Now lets find how there are outliers in the dataset for each type of the propety type vs hability score also property type and property are. This is done just to understand what are the outliers in each property type and their median. THis is achieved using boxplot

```python
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")
sns.boxplot(x = 'Property_Type', y = 'Habitability_score', data = data)
```

```python
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")
sns.boxplot(x = 'Property_Type', y = 'Property_Area', data = data)
```
