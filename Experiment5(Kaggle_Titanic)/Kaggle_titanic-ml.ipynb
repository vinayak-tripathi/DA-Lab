```python
train = pd.read_csv("/kaggle/input/titanic/train.csv",index_col=[0])
```

```python
train.isnull().sum()
```

```python
train.Age.fillna(train.Age.mean(),inplace=True)
train.Embarked.fillna(train.Embarked.mode()[0],inplace=True)
train.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
X = train.drop(columns = ['Cabin','Name','Ticket','Survived'],axis=1)
Y = train['Survived']
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X, Y)
```

```python
test = pd.read_csv("/kaggle/input/titanic/test.csv",index_col=[0])
test.Age.fillna(test.Age.mean(),inplace=True)
test.Fare.fillna(test.Fare.mean(),inplace=True)
test.Embarked.fillna(test.Embarked.mode()[0],inplace=True)
test.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
test.drop(columns = ['Cabin','Name','Ticket'],axis=1,inplace=True)
```

```python
ans = model.predict(test)
```

```python
pd.DataFrame(ans,index = test.index,columns = ['Survived']).to_csv("test_res.csv")
```

### Interpretation
1.  We can say that the Data Has Some hidden trend which is found using Logistic Regression
2. The Data presented was having some missing values. So For ticket_ID and name the attributes were dropped.
3. The Missing Ages were replaced by the mean of the Age. The missing gender and Emarkment were replaced by the Mode of the respective dataset
4. Then since the Logistic Regression needs numerical values so we need to Label encode the attributes with the object datatype and replace them with numerical value
5. We needed to train the model and then predict based on the unseen data. 
6. We got an Accuracy of 76% on the test Dataset
